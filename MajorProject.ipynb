```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(ttnc['year'],ttnc['selling_price'],alpha=0.5)
plt.title('SALES OF CARS IN DIFFERENT YEARS')
plt.xlabel('year')
plt.ylabel('selling_price')
plt.show()
```

```python
ax=ttnc.groupby(['km_driven']) ['selling_price'].mean().plot.bar(
    figsize=(10,5),
    fontsize=14
)
ax.set_title("Average of km_driven and selling_price")
ax.set_xlabel("km_driven",fontsize=15)
ax.set_ylabel("selling_price",fontsize=15)
```
